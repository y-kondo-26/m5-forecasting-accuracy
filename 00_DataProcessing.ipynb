```python
import pandas as pd
import gc
import pickle
import gzip
```

```python
df = pd.read_csv('sales_train_validation.csv')
```

```python
df = df.drop(['id', 'dept_id', 'state_id'], axis=1)
```

```python
df_melt = pd.melt(df, id_vars=['item_id','cat_id','store_id']) # 'dept_id','state_id'
```

```python
del df
gc.collect()
```

```python
df_melt = df_melt.rename(columns={'variable': 'd', 'value': 'sales'})
```

```python
df_melt.head(5)
```

```python
df_melt.describe()
```

```python
df_melt['sales'] = df_melt['sales'].astype('int16')
```

```python
df_melt.head(5)
```

```python
df_analysis = df_melt
```

```python
del df_melt
gc.collect()
```

```python
df_calender = pd.read_csv('calendar.csv')
```

```python
df_sales_cal = pd.merge(df_analysis, df_calender, on='d', how='inner')
```

```python
del df_analysis, df_calender
gc.collect()
```

```python
df_sales_cal.head(5)
```

```python
df_sales_cal = df_sales_cal.drop(['d', 'weekday'], axis=1)
```

```python
df_sales_cal['wk'] = df_sales_cal['wm_yr_wk'].astype(str).str[-2:]
```

```python
df_sales_cal = df_sales_cal.drop(['date'], axis=1)
```
